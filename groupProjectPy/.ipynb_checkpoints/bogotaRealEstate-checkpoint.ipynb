```python
remodelled_boolean = pd.unique(df_loader['remodelado'])
remodelled_boolean = list(remodelled_boolean)

laundry_area_boolean = list(pd.unique(df_loader['zona_de_lavanderia']))

gas_boolean = list(pd.unique(df_loader['gas']))

house_apartment = list(pd.unique(df_loader['tipo_de_inmueble']))

parking_boolean = list(pd.unique(df_loader['parqueadero']))

reception = list(pd.unique(df_loader['porteria']))

locality = list(pd.unique(df_loader['localidad']))

barrio = list(pd.unique(df_loader['barrio']))

estrato = list(pd.unique(df_loader['estrato']))

def encode(uncoded_data):
    '''
        Changes the string values in the data frame to integer values 
        Parameter: 
            uncoded_data - Unique list of values in a dataframe column
        Returns: 
            Dictionary of string to integer value         
    '''
    uncoded_encoded_mapping = {}
    for x in uncoded_data:
        uncoded_encoded_mapping[x] = uncoded_data.index(x)
    return uncoded_encoded_mapping

corresponding_remodelled = encode(remodelled_boolean)
corresponding_laundry = encode(laundry_area_boolean)
corresponding_gas = encode(gas_boolean)
corresponding_property_type = encode(house_apartment)
corresponding_parking = encode(parking_boolean)
corresponding_recption = encode(reception)
correponding_locality = encode(locality)
correponding_barrio = encode(barrio)
correponding_estrato = encode(estrato)

print(corresponding_remodelled)
print(corresponding_laundry)
print(corresponding_gas)
print(corresponding_property_type)
print(corresponding_parking)
print(corresponding_recption)
print(correponding_locality)
print(correponding_estrato)
```

```python
# changing endangered and specie columns to quantifiable values
#https://www.geeksforgeeks.org/using-dictionary-to-remap-values-in-pandas-dataframe-columns/
# https://stackoverflow.com/questions/29576430/shuffle-dataframe-rows
df_quant = df_loader.copy()
def quantifyColumns(column,translation):
    '''
    Mapping the integer values to the corresponding string and updating df
    paramter 
    Parameters:
        column - Dataframe column to update 
        translation - Dictionary of corresponding string:int
    Returns:
        Dataframe with no string values
    '''
    df_quant[column] = df_quant[column].map(translation)
     
    
quantifyColumns("remodelado",corresponding_remodelled)
quantifyColumns("zona_de_lavanderia",corresponding_laundry)
quantifyColumns("tipo_de_inmueble",corresponding_property_type)
quantifyColumns("gas",corresponding_gas)
quantifyColumns("parqueadero",corresponding_parking)
quantifyColumns("porteria",corresponding_recption)
quantifyColumns('localidad',correponding_locality)
quantifyColumns('barrio',correponding_barrio)
quantifyColumns('estrato',correponding_estrato)
data_frame_collection["df_quant"] = df_quant
df_quant.head()

```

```python
def showViolinPlot(data_frame,x_axis,y_axis):
    '''
        Displaying violin plot using seaborn
        Parameters:
            Dataframe - contains the desired data
            x-axis - Column from said dataframe to be plotted on x-axis
            y-axis - Column from said dataframe to be plotted on y-axis
        Returns:
            Nothing, but shows the plot
    '''
    sns.violinplot(x=data_frame[x_axis], y = data_frame[y_axis])
    plt.xticks(
    rotation=25, 
    horizontalalignment='right',
    fontweight='light',
      
)
    plt.show()
showViolinPlot(df_quant,'estrato','price_m2')
#showViolinPlot(df_quant,'habitaciones','price_m2')
#showViolinPlot(df_quant,'estrato','price_m2')
```

```python
input_shape_0 = len(list(X.columns))
print(input_shape_0)
print(len(estrato))
```

```python
from keras.layers import Flatten

def model_builder(hp):
    '''
        Outline of the model, defining the options for the keras tuner
        Parameters:
            hp - Internally passed object. nstantiates the Hyperparameter
                    object of the keras tuner
        Returns: 
            Building blocks for model
    '''
    model = Sequential()
    model.add(Flatten(input_shape = (input_shape_0,)))
   # hp_activation = hp.Choice('activation', values = ['LeakyReLU','elu','gelu','tanh','sigmoid']) 
    hp_activation = hp.Choice('activation', values = ['LeakyReLU','elu','gelu']) 
    hp_layer_1 = hp.Int('layer_1', min_value = 0, max_value = 512, step = 16)
    hp_layer_2 = hp.Int('layer_2', min_value = 0, max_value = 512, step = 16)
    hp_layer_3 = hp.Int('layer_3', min_value = 0, max_value = 512, step = 16)
  #  hp_layer_4 = hp.Int('layer_4', min_value = 0, max_value = 512, step = 16)
    hp_learning_rate = hp.Choice('learning_rate', values = [0.01, 0.001, 0.0001])
    hp_rho = hp.Choice('rho', values = [0.98,0.9,0.97,0.96])
 #   hp_momentum = hp.Choice('momentum', values = [0.0,0.001,0.01])
  #  hp_beta1 = hp.Choice('beta_1', values = [0.9,0.95,0.99])
   # hp_beta2 = hp.Choice('beta_2', values = [0.995,0.999,0.9999])
    
    
    model.add(Dense(units = hp_layer_1, activation = hp_activation))
    
    model.add(Dense(units = hp_layer_2, activation = hp_activation, kernel_regularizer=regularizers.L1L2(l1=1e-5, l2=1e-4),
    bias_regularizer=regularizers.L2(1e-4),
    activity_regularizer=regularizers.L2(1e-5)))
    
    model.add(Dense(units = hp_layer_3, activation = hp_activation, kernel_regularizer=regularizers.L1L2(l1=1e-5, l2=1e-4),
    bias_regularizer=regularizers.L2(1e-4),
    activity_regularizer=regularizers.L2(1e-5)))
  #  model.add(Dense(units = hp_layer_4, activation = hp_activation, kernel_regularizer=regularizers.L1L2(l1=1e-5, l2=1e-4),
   # bias_regularizer=regularizers.L2(1e-4),
    #activity_regularizer=regularizers.L2(1e-5)))
    if(prediction_column == 'price_m2'):
        model.add(Dense(1, activation='linear'))
        model.compile(optimizer = keras.optimizers.Nadam(learning_rate=hp_learning_rate), 
                  loss='mean_squared_error', 
                  metrics = ['mean_absolute_percentage_error'] )
    else: 
        model.add(Dense(len(estrato), activation = 'softmax'))
        model.compile(optimizer = keras.optimizers.RMSprop(learning_rate=hp_learning_rate, rho=hp_rho),                  
                      loss='sparse_categorical_crossentropy', 
                      metrics = ['accuracy'] )
    return model
```

```python
# better than hyperband
objective_value = ''
monitor_value = ''
trials = -1
restoreFlag = False
patience_value = -1
direction = ''
if prediction_column == 'price_m2':
    objective_value = 'mean_absolute_percentage_error'
    monitor_value = 'mean_absolute_percentage_error'
    trials = 12
    patience_value = 4
    validation_value = 0.18

else:
    objective_value = 'val_accuracy'
    monitor_value = 'val_loss'
    trials = 20
    restoreFlag = True
    patience_value = 4
    
tuner = kt.RandomSearch(
    model_builder,
    objective = objective_value,
    max_trials=trials,
    directory='dir',
    overwrite=True,
    project_name='bogot'
)
```

```python
es = keras.callbacks.EarlyStopping(
                                   #monitor='val_loss',
                                   monitor = monitor_value,
                                   mode='auto',
                                   patience=patience_value, 
                                   restore_best_weights=restoreFlag)
```

```python
tuner.search(X_train, y_train, epochs = 100,
            validation_split = 0.18,
            callbacks = [es])
```

```python
hyper_parameters = tuner.get_best_hyperparameters(num_trials = 1)[0]
model = tuner.hypermodel.build(hyper_parameters)
print(model)
history = model.fit(X_train, y_train, epochs = 120, 
                    validation_split = 0.18, 
                    shuffle=True,
                    callbacks = [es])
```

```python
for k,v in hyper_parameters.values.items():
    print(f'{k} : {v}')
```

```python
history_dict = history.history
print(history_dict.keys())
if(prediction_column == 'price_m2'):
    mape = history_dict['mean_absolute_percentage_error']
    val_mape = history_dict['val_mean_absolute_percentage_error']
    epochs = range(1, len(mape) + 1)
 #   print(len(mape))
  #  print(len(val_mape))
    plt.plot(epochs, val_mape, label='Validation mape', color='blue')
    plt.plot(epochs, mape, label='Training mape', color='orange')
    plt.title('Training and validation mape')
    plt.xlabel('Epochs')
    plt.ylabel('mape')
    plt.legend()
    plt.show()

else:
    # accuracy
    acc = history_dict['accuracy']
    val_acc = history_dict['val_accuracy']

    # loss
    loss = history_dict['loss']
    val_loss = history_dict['val_loss']

    # range of X (no. of epochs)
    epochs = range(1, len(acc) + 1)

    # accuracy plot

    plt.plot(epochs, val_acc, label='Validation accuracy', color='blue')
    plt.plot(epochs, acc, label='Training accuracy',color='orange')
    plt.title('Training and validation accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    plt.plot(epochs, loss, label='Loss', color = 'red')
    plt.plot(epochs, val_loss, label='Validation Loss', color = 'purple')
    plt.title('Training and validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```
